```python
df = pd.read_sql_query(
    """
    SELECT customer_id, sum(amount) 
    FROM payment
    WHERE customer_id = 1
    OR customer_id = 4
    GROUP BY customer_id;
    """
, conn)

df 
```

```python
# Max

df = pd.read_sql_query(
    """
    SELECT p.customer_id, max(p.amount) 
    FROM payment p
    WHERE p.customer_id = 1;
    
    """
    , conn)

df
```

```python
# Min

df = pd.read_sql_query(

    """
    SELECT p.customer_id, min(p.amount) 
    FROM payment p
    WHERE p.customer_id = 1;
    
    """
    , conn)

df
```
